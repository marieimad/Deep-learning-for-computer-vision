````python
# ── 3. Download How2Sign annotations (CSV) ────────────────────
# Les annotations sont libres d'accès
import urllib.request

ANN_URLS = {
    'train': 'https://how2sign.github.io/data/how2sign_realigned_train.csv',
    'val':   'https://how2sign.github.io/data/how2sign_realigned_val.csv',
    'test':  'https://how2sign.github.io/data/how2sign_realigned_test.csv',
}

for split, url in ANN_URLS.items():
    dst = f'{ROOT}/annotations/{split}.csv'
    if not os.path.exists(dst):
        print(f'Downloading {split}...')
        urllib.request.urlretrieve(url, dst)
    else:
        print(f'✅ {split}.csv déjà présent')
````

````python
# ── 4. Vérifier les annotations ───────────────────────────────
import pandas as pd

df_train = pd.read_csv(f'{ROOT}/annotations/train.csv', sep='\t')
df_val   = pd.read_csv(f'{ROOT}/annotations/val.csv',   sep='\t')
df_test  = pd.read_csv(f'{ROOT}/annotations/test.csv',  sep='\t')

print(f'Train : {len(df_train)} exemples')
print(f'Val   : {len(df_val)} exemples')
print(f'Test  : {len(df_test)} exemples')
print()
print('Colonnes :', df_train.columns.tolist())
print()
df_train.head(3)
````
